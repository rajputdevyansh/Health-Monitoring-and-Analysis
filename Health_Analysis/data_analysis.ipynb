```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
data=pd.read_csv('preprocessed_data.csv')
```

```python
# Viewing the data
data.head()
```

```python
# Plotting the distribution of the Age column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['Age'],bins=20,kde=True,color='red')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of HeartRate column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['HeartRate'],bins=20,kde=True,color='blue')
plt.title('Distribution of Heart Rate')
plt.xlabel('HeartRate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of RespiratoryRate column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['RespiratoryRate'],bins=20,kde=True,color='green')
plt.title('Distribution of Respiratory Rate')
plt.xlabel('RespiratoryRate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of BodyTemperature column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['BodyTemperature'],bins=20,kde=True,color='purple')
plt.title('Distribution of Body Temperature')
plt.xlabel('BodyTemperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of OxygenSaturation column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['OxygenSaturation'],bins=20,kde=True,color='Indigo')
plt.title('Distribution of Oxygen Saturation')
plt.xlabel('OxygenSaturation')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the distribution of SystolicPressure and DiastolicPressure columns
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.histplot(data['SystolicPressure'],bins=20,kde=True,color='red',label='SystolicPressure')
sns.histplot(data['DiastolicPressure'],bins=20,kde=True,color='orange',label='DiastolicPressure')
plt.title('Distribution of Blood Pressure')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plotting gender distribution
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
gender_counts=data['Gender'].value_counts()
gender_counts.plot.pie(autopct='%1.1f%%',colors=['Teal','Tan'],shadow=True,explode=(0.1,0))
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Plotting the distribution of the Age Groups column
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.countplot(data['ACategory'])
plt.title('Distribution of ACategory')
plt.xlabel('ACategory')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting correlation matrix
matrix_data=data[['HeartRate','RespiratoryRate','BodyTemperature','OxygenSaturation','SystolicPressure','DiastolicPressure']].corr()
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.heatmap(matrix_data.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# ploting HeartRate by ActivityLevel
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.boxplot(x='ActivityLevel',y='HeartRate',data=data,palette='coolwarm',hue='ActivityLevel')
plt.title('Heart Rate by Activity Level')
plt.show()
```
